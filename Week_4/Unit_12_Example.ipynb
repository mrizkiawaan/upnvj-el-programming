##Example

```python
# Import pandas library
import pandas as pd

data = [4, 9, 2, 7, 1, 5, 8, 3, 6, 10]

df = pd.DataFrame(data, columns=['numbers'])
df
```

```python
df_mean = df['numbers'].mean()
df_median = df['numbers'].median()
df_std = df['numbers'].std()

print("mean of all dataset is :", df_mean)
print("median of all dataset is :", df_median)
print("std of all dataset is :", df_std)
```

###Jika yang dimaksud cari mean, median, std satu kolom

```python
# Adding result to dataset
df['dataset_mean'] = df_mean
df['dataset_median'] = df_median
df['dataset_std'] = df_std
```

```python
# Print dataset
df
```

###Jika yang dimaksud mean, median, std setiap baris

```python
data_mean = []
data_median = []
data_std = []

for ind in df.index:
  data_mean.append(df['numbers'].iloc[ind].mean())
  # data_median.append(df['numbers'].iloc[ind].median()) # this line result is error
  data_std.append(df['numbers'].iloc[ind].std())
```

```python
# Adding result to dataset
df['row_mean'] = data_mean
# df['row_median'] = data_median
df['row_std'] = data_std
```

##Another example

```python
sample_df = pd.read_csv("kpop_sample_dataset.csv")
sample_df
```

```python
city = sample_df.iloc[:, 4:7]
city
```

```python
# Compute mean,median,std for each row
fans_mean_val = city.mean(axis=1)
fans_median_val = city.median(axis=1)
fans_std_val = city.std(axis=1)
```

```python
# Add mean,median,std to df
sample_df['fans_mean'] = fans_mean_val
sample_df['fans_median'] = fans_median_val
sample_df['fans_std'] = fans_std_val
```

```python
pd.options.display.float_format = '{:.2f}'.format
sample_df
```

```python
# create figure and subplots
sns.set_palette("Spectral")
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(30, 5))

# plot on first subplot
# sns.displot(df, x="sepal length (cm)", kde=True,ax = axs[0]) 
sns.histplot(sample_df['Bangkok'], kde=True,ax = axs[0])
axs[0].set_title('distribution plot')

# plot on second subplot
sns.scatterplot( y = 'Songs', x = 'Duration', data = sample_df, ax = axs[1])
axs[1].set_title('scatterplot')

# plot on third subplot
sns.boxplot( y = 'Songs', x = 'Members', data = sample_df, ax = axs[2])
axs[2].set_title('boxplot')

# set overall title and show plot
plt.suptitle('Example data visualization',size = 25)
plt.show()
```
