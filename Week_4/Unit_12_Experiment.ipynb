###Create new data from analytics of example dataset

```python
# Assign all variable for sepal length
sepal_length_mean = df['sepal length (cm)'].mean(); sepal_length_median = df['sepal length (cm)'].median(); sepal_length_std = df['sepal length (cm)'].mean()

# Assign all variable for sepal width
sepal_width_mean = df['sepal width (cm)'].mean(); sepal_width_median = df['sepal width (cm)'].median(); sepal_width_std = df['sepal width (cm)'].std()

# Assign all variable for petal length
petal_length_mean = df['petal length (cm)'].mean(); petal_length_median = df['petal length (cm)'].median(); petal_length_std = df['petal length (cm)'].mean()

# Assign all variable for petal width
petal_width_mean = df['sepal width (cm)'].mean(); petal_width_median = df['sepal width (cm)'].median(); petal_width_std = df['sepal width (cm)'].std()
```

```python
# Create list to contain data from looping
mean_sepal_length_values = [] # Contains values calculated by adding the mean sepal length to each data value in the "sepal length" column
median_sepal_length_values = [] # Contains values calculated by adding the median sepal length to each data value in the "sepal length" column
std_dev_sepal_length_values = [] # Contains values calculated by adding the standard deviation of sepal length to each data value in the "sepal length" column

mean_sepal_width_values = [] # Contains values calculated by adding the mean sepal width to each data value in the "sepal width" column
median_sepal_width_values = [] # Contains values calculated by adding the median sepal width to each data value in the "sepal width" column
std_dev_sepal_width_values = [] # Contains values calculated by adding the standard deviation of sepal width to each data value in the "sepal width" column

mean_petal_length_values = [] # Contains values calculated by adding the mean petal length to each data value in the "petal length" column
median_petal_length_values = [] # Contains values calculated by adding the median petal length to each data value in the "petal length" column
std_dev_petal_length_values = [] # Contains values calculated by adding the standard deviation of petal length to each data value in the "petal length" column

mean_petal_width_values = [] # Contains values calculated by adding the mean petal width to each data value in the "petal width" column
median_petal_width_values = [] # Contains values calculated by adding the median petal width to each data value in the "petal width" column
std_dev_petal_width_values = [] # Contains values calculated by adding the standard deviation of petal width to each data value in the "petal width" column
```

```python
# Add all calculated value to list
for ind in df.index:
  mean_sepal_length_values.append(df["sepal length (cm)"].values[ind] + sepal_length_mean)
  median_sepal_length_values.append(df["sepal length (cm)"].values[ind] + sepal_length_median)
  std_dev_sepal_length_values.append(df["sepal length (cm)"].values[ind] + sepal_length_std)

  mean_sepal_width_values.append(df["sepal width (cm)"].values[ind] + sepal_width_mean)
  median_sepal_width_values.append(df["sepal width (cm)"].values[ind] + sepal_width_median)
  std_dev_sepal_width_values.append(df["sepal width (cm)"].values[ind] + sepal_width_std)

  mean_petal_length_values.append(df["petal length (cm)"].values[ind] + petal_length_mean)
  median_petal_length_values.append(df["petal length (cm)"].values[ind] + petal_length_median)
  std_dev_petal_length_values.append(df["petal length (cm)"].values[ind] + petal_length_std)

  mean_petal_width_values.append(df["petal width (cm)"].values[ind] + petal_width_mean)
  median_petal_width_values.append(df["petal width (cm)"].values[ind] + petal_width_median)
  std_dev_petal_width_values.append(df["petal width (cm)"].values[ind] + petal_width_std)
```

```python
# Assign all list to one list to create dataframe
df_analytics_res = {
  'sepal length mean' : mean_sepal_length_values,
  'sepal length median' : median_sepal_length_values,
  'sepal length std' : std_dev_sepal_length_values,

  'sepal width mean' : mean_sepal_width_values,
  'sepal width median' : median_sepal_width_values,
  'sepal width std' : std_dev_sepal_width_values,

  'petal length mean' : mean_petal_length_values,
  'petal length median' : median_petal_length_values,
  'petal length std' : std_dev_petal_length_values,

  'petal width mean' : mean_petal_width_values,
  'petal width median' : median_petal_width_values,
  'petal width std' : std_dev_petal_width_values,
}
```

```python
analytics_df = pd.DataFrame(df_analytics_res)
```
