##1.Fungsi Grafik Eksponen

```python
# import numpy as np
# import matplotlib.pyplot as plt
# Buatlah nilai x dengan interval [0, 2π] dan step π/100
x = np.arange(0, 2*np.pi, np.pi/100)
# Masukkan fungsi y1 = sin(x) dan y2 = cos(x + 2π)
y1 = np.sin(x)
y2 = np.cos(x+np.pi)
plt.plot(x, y1, '*', label='Sin(x)')
plt.plot(x, y2, '+', label='cos(x + 2π)')
plt.xlabel('Sumbu-X')
plt.ylabel('Sumbu-y')
plt.title('Plot Data')
plt.grid()
plt.legend()
plt.show()
```

##2.Mencari nilai dari gelombang segitiga dengan menggunakan deret Fourrier

```python
# import numpy as np
# import matplotlib.pyplot as plt
t = np.linspace(0 , 10 , 1000)
plt.axvline(x = 0, color = 'k')
plt.axhline(y = 0, color = 'k')
plt.xlabel('Time (s)')
plt.ylabel('f(t)')
sum = 0
amplitudo = int(input("Masukkan nilai amplitudo: "))
periode = int(input("Masukkan nilai T : "))
n = int(input("Masukkan banyaknya aproksimasi : "))
for angka in range(1, n):
# Masukkan fourier transform di bawah
  sum = sum + amplitudo * np.sin(2 * np.pi * angka * t / periode)
# Masukkan nilai t sebagai x dan Sum sebagai Y
plt.plot(t, sum)
```

##3. Mencari nilai kompleks pada suatu rangkaian menggunakan numpy

! Modal browsing di internet, hitung lagi untuk memastikan

```python
# import numpy as np
# Diketahui v(t) = 50∠30, L = 20.10^(-3), C = 50.10^(-6)
# Masukkan nilai C, L dan f
f = 50 # in Hz
C = 50e-6 # in Farads
L = 20e-3 # in Henrys
W = 2*np.pi*f # in radians
R = 0 # in ohms
# Menghitung nilai masing-masing impendansi
# Ingat ZL = 1j*W*L, ZC = -1j/(W*C), ZR = R
ZL = 1j*W*L
ZC = -1j/(W*C)
ZR = R
def ke_kompleks(amplitudo, sudut):
  # A∠θ = A(cos(θ) + j*sin(θ))
  real = amplitudo * np.cos(sudut * (np.pi/180))
  imajiner = 1j*amplitudo * np.sin(sudut * np.pi/180)
  # Tambahkan bilangan real dan imajiner
  bilangan_kompleks = real + imajiner
  return bilangan_kompleks

def ke_polar(fungsi_kompleks):
  # Fungsi kompleks = A(cos(θ) + j*sin(θ))
  #Amplitudo (A) = akar(real^2(fungsi_kompleks) + imajiner^2(fungsi_kompleks))
  # Sudut(θ) = tan^(-1)(bilangan_imajiner/bilangan_real)
  amplitudo = np.sqrt(np.real(fungsi_kompleks)**2 + np.imag(fungsi_kompleks)**2)
  sudut = np.angle(fungsi_kompleks, deg=True)
  # Membulatkan 2 digit dibelakang 0
  bilangan_polar = str(round(amplitudo, 2)) + '∠' + str(round(sudut, 2))
  return bilangan_polar
# Ztot = ZL + ZC + ZR
Ztot = ZL + ZC + ZR
# Mengubah v ke dalam kompleks
v = ke_kompleks(50, 30)
# Arus = v / ztot
arus = v / Ztot
arus_dalam_polar = ke_polar(arus)
print(f"Nilai arus dalam polar adalah : {arus_dalam_polar}")
```

##4. Buatlah fungsi 3d dari z = f(x,y) = x2 + y2

```python
# import numpy as np
# import matplotlib.pyplot as plt
ax = plt.axes(projection='3d')
# Buatlah interval dari sumbu x dan y [-100,100] dengan kenaikan per 100
xn = np.linspace(-100, 100, 100)
yn = np.linspace(-100, 100, 100)
X, Y = np.meshgrid(xn, yn)
ax.set_xlabel("Sumbu-X")
ax.set_ylabel("Sumbu-Y")
ax.set_title("Z = f(x,y) = x^2 + y^2")
# Buatlah fungsi Z = x^2 + y^2 dari nilai yang telah di meshgrid
Z = X**2 + Y**2 # ** artinya kuadrat
ax.plot_surface(X, Y, Z,cmap="jet")
plt.show()
```
