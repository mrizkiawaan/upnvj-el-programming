```python
# Import all module required
import sympy as sym
```

```python
t = sym.symbols('t')
s = sym.symbols('s')
# define the function to be transformed
fungsi = 3*t**2
# apply Laplace transformation to the function
fungsi_laplace = sym.laplace_transform(fungsi, t, s, noconds=True)
# print the Laplace-transformed function
print(fungsi_laplace)
```

```python
t = sym.symbols('t')
a = sym.symbols('a')
k = sym.symbols('k')
fungsi = sym.exp(-a*t)*sym.Heaviside(t)
sym.fourier_transform(fungsi, t, k)
```

```python
# y(n) = x(n)*h(n)
import scipy.signal as signal
import numpy as np
x = np.array([-1,2,0,1])
h = np.array([3,1,0,-1])
y = signal.convolve(x, h)
print(f"Nilai konvolusi dari y(n) = x(n)*h(n) adalah : \n{y}")
```
